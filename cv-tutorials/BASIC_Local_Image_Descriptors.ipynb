***Harris corner detector***

- The Harris corner detection algorithm (or sometimes the Harris & Stephens corner detector) is one of the simplest corner indicators available
- The general idea is to locate interest points where the surrounding neighborhood shows edges in more than one direction, these are then image corners
- Implemented in skimage.feature package - three step process: 
    - (1) use corner detector to find `response` image, 
    - (2) use `corner_peaks` or `peak_local_max` to find coordinates or boolean array of corners
    - (3) OPTIONALLY, use `corner_subpix` to determine the subpixel corner positions 
- **NOTE that the corners found are in [[row, col]...] format, which corresponds to [[y, x], ...] coordidate in drawing**

```python
%pylab inline
```

```python
from skimage import color, io, feature

## Original Image
img = color.rgb2gray(io.imread('data/python-cv/empire.jpg'))
## Harris corner response
harris_response = feature.corner_harris(img, sigma = 1.5, )
## corners with different thresholds
## NOTE THAT rows -- ys and cols -- xs
ys_small, xs_small = zip(*feature.corner_peaks(harris_response, threshold_rel=0.01, ))
ys_big, xs_big = zip(*feature.corner_peaks(harris_response, threshold_rel=0.1))
## Plot images
fig, axes = subplots(nrows = 2, ncols = 2, figsize=(12, 12))
gray()
axes[0, 0].imshow(img)
axes[0, 0.].set_title('original image')
axes[0, 1].imshow(harris_response)
axes[0, 0.].set_title('harris response')
axes[1, 0].imshow(img)
axes[1, 0].plot(xs_small, ys_small, 'b+')
axes[1, 0.].set_title('harris corners $thr=0.01$')
axes[1, 0].axis('off')
axes[1, 1].imshow(img)
axes[1, 1].plot(xs_big, ys_big, 'b+')
axes[1, 1.].set_title('harris corners $thr=0.1$')
axes[1, 1].axis('off')
```

***Decriptors for Corners***

- The Harris corner detector gives interest points in images but does not contain an inherent way of comparing these interest points across images to find matching corners.
- What we need is to add a descriptor to each point and a way to compare such descriptors.
- An **interest point descriptor** is a vector assigned to an interest point that describes the image appearance around the point.
- With point correspondence or corresponding points we mean points in different images that refer to the same object or scene point. The better the descriptor, the better your correspondences will be.

***Robust matching using RANSAC***

- Steps: 
    - (1) Find interest points in two correlated images (based on harris descriptors)
    - (2) Find correspondences based on weighted sum of squared differences of a small neighborhood around them (patches)
    - (3) After finding the correspondences we end up having a set of source and destination coordinates which can be used to `estimate the geometric transformation` between both images
    - (4) Therefore, the RANSAC algorithm is used on top of the normal model to robustly estimate the parameter set by detecting outliers.
-  Note, that this measure is only robust towards linear radiometric and not geometric distortions and is thus only usable with slight view point changes.

Examples from http://scikit-image.org/docs/dev/auto_examples/plot_matching.html

```python
## download images
import json, urllib, urlparse, os
from os import path
images_url = 'http://www.panoramio.com/map/get_panoramas.php?order=popularity&set=public&from=0&to=20&minx=-77.037564&miny=38.896662&maxx=-77.035564&maxy=38.898662&size=medium'
conn = urllib.urlopen(images_url)
content = json.loads(conn.read())
imurls = [im['photo_file_url'] for im in content['photos']]
## download images
if not path.exists('data/python-cv/downloaded'):
    os.mkdir('data/python-cv/downloaded')
for imurl in imurls:
    filename = path.join('data/python-cv/downloaded', path.basename(urlparse.urlparse(imurl).path))
    img = urllib.URLopener()
    img.retrieve(imurl, filename)
    print 'downloading:', imurl
```

```python
from IPython.parallel import Client
client = Client()
print len(client)
dv = client[:]
dv.block = True
dv.execute('from functools import partial')
```

```python
## get DAISY descriptors
%%px
def get_daisy_descriptors(img, step=180, radius=58):
    from skimage import color, feature
    img = color.rgb2gray(img)
    descs = feature.daisy(img, step = step, radius = radius)
    return descs.flatten()
```

```python
img = io.imread('data/python-cv/downloaded/5752686.jpg')
descs = get_daisy_descriptors(img)
print descs.shape
```

```python

whitehouses = [io.imread(path.join('data/python-cv/downloaded/', f)) 
        for f in os.listdir('data/python-cv/downloaded/')]

desc_vectors = dv.map(get_daisy_descriptors, whitehouses)
```

```python
print [v.shape for v in desc_vectors]
wh_X = np.vstack(desc_vectors)
print wh_X.shape
```
