### General Pattern
- `cv2.FeatureDetector_create()` to create feature detectors (which `detect` keypoints), choices are "FAST", "STAR", "SIFT", "SURF", "ORB", "BRISK", "MSER", "GFTT", "HARRIS", "Dense", "SimpleBlob". Also a combined format is supported: feature detector adapter name ( "Grid" – GridAdaptedFeatureDetector, "Pyramid" – PyramidAdaptedFeatureDetector ) + feature detector name (see above), for example: "GridFAST", "PyramidSTAR". See the [doc](http://docs.opencv.org/modules/features2d/doc/common_interfaces_of_feature_detectors.html)
- `cv2.DescriptorExtractor_create()` to create feature extractors (which extract feature vectors from keypoints). Parameters are "BRISK" and "ORB". see the [doc](http://docs.opencv.org/trunk/modules/features2d/doc/common_interfaces_of_descriptor_extractors.html)

### Finding SIFT Features
- invariant to both scalar and oritantion
- `cv2.SIFT().detect()` takes an gray image and search for keypoints in the image, each of which is a special structure which has many attributes like its **(x,y), neighborhood size, orientation, and strenght**.
- use cv2.drawKeypoints to draw those keypoints on the image
- after getting keypoints, use `SIFT.compute()` to get the descriptors, e.g., by kp, des = SIFT.compute(gray, kp)
- or two steps in one: `kp, des = SIFT.detectAndCompute(gray, None)`

***Opencv 3.0 has removed SIFT and SURF from the default package***

### Finding SURF Features
- faster version of SIFT, with slight accuracy loss

***Opencv 3.0 has removed SIFT and SURF from the default package***

### FAST (Features from Accelerated Segment Test) algoirthm for Corner Detection
- It got speeded up by machine learning, e.g., ID3 decision tree
- It is several times faster than other existing corner detectors.
- But it is not robust to high levels of noise. It is dependant on a threshold.
- BUT it doesnt give descriptions!

```python
#img = cv2.imread("public-images/fish.png")[:,:,::-1]
img = cv2.imread("public-images/coins.jpg")
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap = plt.cm.gray)

fast = cv2.FastFeatureDetector()
kp = fast.detect(gray_img, None)
fast_img = cv2.drawKeypoints(gray_img, kp, None, color = (0, 255, 0), )
plt.imshow(fast_img)
```

### BRIEF (Binary Robust Independent Elementary Features)
- SIFT and SURF generates high dimensional **float features**, In actual matching, these features may need to be compressed by PCA, LDA and LSH, to get **binary strings**.
- then those **binary strings** can be compared and matched using Hamming distance.
- BRIEF provides a shortcut to find **binary string features** directly without finding descriptors first.
- One important point is that BRIEF is a **feature descriptor**, it doesn’t provide any method to find **the features**. So you will have to use any other feature detectors like SIFT, SURF etc.The paper recommends to use CenSurE which is a fast detector and BRIEF works even slightly better for CenSurE points than for SURF points.

```python
## FAST + BRISK -> the only way that works

## fast detector
fast = cv2.FastFeatureDetector()
kp = fast.detect(gray_img, None)

## BRIEF extractor
brief = cv2.DescriptorExtractor_create("BRISK")
## compute the descriptors - enhacned kp information
kp, des = brief.compute(gray_img, kp)

print des.shape, des.dtype

kp_img = cv2.drawKeypoints(gray_img, kp, None)
plt.imshow(kp_img, cmap = plt.cm.gray)
```

```python
## SIFT doesnt work

## STAR detector
star = cv2.FeatureDetector_create("FAST", )
## find feature keppoints
kp = star.detect(gray_img)

## BRIEF extractor
brief = cv2.DescriptorExtractor_create("ORB")
## compute the descriptors
kp, des = brief.compute(gray_img, kp)

print des.shape, des.dtype

kp_img = cv2.drawKeypoints(gray_img, kp, None)
plt.imshow(kp_img, cmap = plt.cm.gray)
```

```python
## STAR doesnt work either

## STAR detector
star = cv2.FeatureDetector_create("BRISK", )
#star = cv2.Feature2D_create("FAST")
## find feature keppoints
kp = star.detect(gray_img)

## BRIEF extractor
brief = cv2.DescriptorExtractor_create("BRISK")
## compute the descriptors
kp, des = brief.compute(gray_img, kp)

print des.shape, des.dtype

kp_img = cv2.drawKeypoints(gray_img, kp, None)
plt.imshow(kp_img, cmap = plt.cm.gray)
```

### ORB (Oriented FAST and Rotated BRIEF)
- as its name suggests, ORB is basically a fusion of FAST keypoint detector and BRIEF descriptor with many modifications to enhance the performance. 
- so its a feature descriptor and can be created using `cv2.DescriptorExtractor_create()`
